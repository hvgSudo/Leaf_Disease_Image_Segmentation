```python
from glob import glob
data = glob('train_images/**/*.jpg', recursive = True)
```

```python
import cv2
import matplotlib.pyplot as plt

for i in data[:5]:
    img = cv2.imread(i)
    img_1 = cv2.resize(img, (200, 200))
    plt.imshow(img_1, cmap = 'binary')
    plt.show()
```

```python
images = []
labels = []

for i in data[:15000]:
    if i.endswith('.jpg'):
        label = i[-5]
        img = cv2.imread(i)
        img_1 = cv2.resize(img, (100, 100))
        images.append(img_1)
        labels.append(label)
```
